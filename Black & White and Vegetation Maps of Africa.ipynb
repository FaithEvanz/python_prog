<h3> Map of Uganda, Lira, DRC and East Africa Map</h3>
<b> black and white and vegetation</b>

```python
#Black and White Map of Uganda
ug_map = folium.Map(location=[1.37,32.29], zoom_start=6, tiles='stamenToner')
ug_map
```

```python
#Vegetation and water bodies Map of Uganda
ug_map = folium.Map(location=[1.37,32.29], zoom_start=6, tiles='stamen Terrain')
ug_map
```

```python
#Black and White Map of Lira District
lira_map = folium.Map(location=[2.25,32.88], zoom_start=9, tiles='stamenToner')
lira_map
```
